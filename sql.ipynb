```python
status = sqlite3.connect("/gdrive/My Drive/colabpractice/dcbot/data/pokemon.db")
win_and_lose =sqlite3.connect("/gdrive/My Drive/colabpractice/dcbot/data/win_and_lose.db")
```

```python
import json
import os
with open(os.path.join("./data/", "pokemon.json"), newline='', encoding='UTF-8') as jsonfile:
    pokemon = json.load(jsonfile)
    jsonfile.close()
with open(os.path.join("./data/", "rank.json"), newline='', encoding='UTF-8') as jsonfile:
    rank = json.load(jsonfile)
    jsonfile.close()

for user in pokemon.keys():
    id=user
    name=pokemon[user][0]
    hisstatus=pokemon[user][1]
    atype=hisstatus[7]
    btype=""
    level=hisstatus[6]
    hp=hisstatus[0]
    speed=hisstatus[5]
    atk=hisstatus[1]
    defv=hisstatus[2]
    spatk=hisstatus[3]
    spdef=hisstatus[4]
    win=rank[user]["win"]
    lose=rank[user]["lose"]
    money=win*100
    exp=win*10
    
    cmd = "insert into pokemon(id	,name,typea,typeb,level,hp,	speed,	atk,	def,	spatk,	spdef) values('{}','{}','{}','{}','{}','{}','{}','{}','{}','{}','{}')".format(id,name,atype,btype,level,hp,speed,atk,defv,spatk,spdef)
    status.execute(cmd)
    status.commit()
    cmd = "insert into wl(id	,win,lose,money,exp) values('{}','{}','{}','{}','{}')".format(id,win,lose,money,exp)
    win_and_lose.execute(cmd)
    win_and_lose.commit()
```
